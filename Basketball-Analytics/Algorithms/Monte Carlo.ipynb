```python
# Function to get the Monte Carlo Dataframe

def gdf_monte():
    team_abr = {'Toronto': 'TOR',
                'Boston': 'BOS',
                'Philadelphia': 'PHI',
                'Cleveland': 'CLE',
                'Indiana': 'IND',
                'Miami': 'MIA',
                'Milwaukee': 'MIL',
                'Washington': 'WAS',
                'Detroit': 'DET',
                'Charlotte': 'CHO',
                'New York': 'NYK',
                'Brooklyn': 'BRK',
                'Chicago': 'CHI',
                'Orlando': 'ORL',
                'Atlanta': 'ATL',
                'Houston': 'HOU',
                'Golden State': 'GSW',
                'Portland': 'POR',
                'Oklahoma City': 'OKC',
                'Utah': 'UTA',
                'New Orleans': 'NOP',
                'San Antonio': 'SAS',
                'Minnesota': 'MIN',
                'Denver': 'DEN',
                'L.A. Clippers': 'LAC',
                'L.A. Lakers': 'LAL',
                'Sacramento': 'SAC',
                'Dallas': 'DAL',
                'Memphis': 'MEM',
                'Phoenix': 'PHO'}


    def query_team(tid):
        q1 = "SELECT tid, pts FROM nba_game_data WHERE tid='{}' AND date>20201210;".format(tid)
        return q1


    opp_points = []
    team_points = []
    for team in team_abr.values():
        games_query = "SELECT gid FROM nba_game_data WHERE tid='{}' AND date>20201210;".format(team.lower())
        all_games = read_query(cnxn, games_query)
        for game in all_games:
            gid = game[0]
            query2 = "SELECT pts FROM nba_game_data WHERE tid!='{}' AND gid='{}' AND date>20201210".format(team.lower(),
                                                                                                           gid)
            opp_points.append(read_query(cnxn, query2))
        avg = []
        for opp in opp_points:
            avg.append(opp[0][0])

        q = query_team(team.lower())
        team_points.append(read_query(cnxn, q))

    team = []
    all_team_points = []

    for x in team_points:
        for t in x:
            team.append(t[0])
            all_team_points.append(t[1])

    opp_points_cleaned = []
    for op in opp_points:
        opp_points_cleaned.append(op[0][0])

    print(len(team), len(all_team_points), len(opp_points_cleaned))

    gdf = pd.DataFrame(team)
    print(gdf.head())
    gdf.columns = ['teams']
    
    gdf['team_points'] = all_team_points
    gdf['opp_points'] = opp_points_cleaned
    return gdf
```

```python
# Get Monte Carlo dataframe from the function and set to gdf

import random as rnd
gdf = gdf_monte()
```

```python
## Pick Teams and run Monte Carlo Simulation

h_team='tor'
a_team='gsw'

home_monte = gdf.loc[gdf['teams'] == h_team]
away_monte = gdf.loc[gdf['teams'] == a_team]

meanpts1 = home_monte.team_points.mean()
meanpts2 = away_monte.team_points.mean()
sdpts1 = home_monte.team_points.std()
sdpts2 = away_monte.team_points.std()

meanopp1 = home_monte.opp_points.mean()
meanopp2 = away_monte.opp_points.mean()
sdopp1 = home_monte.opp_points.std()
sdopp2 = away_monte.opp_points.std()

def gameSim():
    GSWScore = (rnd.gauss(meanpts1,sdpts1)+ rnd.gauss(meanopp2,sdopp2))/2
    CLScore = (rnd.gauss(meanpts2,sdpts2)+ rnd.gauss(meanopp1,sdopp1))/2
    if int(round(GSWScore)) > int(round(CLScore)):
        return 1
    elif int(round(GSWScore)) < int(round(CLScore)):
        return -1
    else: return 0

def gamesSim(ns=1000):
    gamesout = []
    team1win = 0
    team2win = 0
    tie = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
        else: tie +=1 
    if team1win >=team2win:
        return str(h_team) + ' is projected to win by ' +str(team1win) 
    else:
        return str(a_team) + ' is projected to win by ' + str(team2win) 

print(gamesSim())
```
