```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os



# Site we will be scraping - get all Bulls game data and put into a dataframe
URL = 'https://www.basketball-reference.com/teams/CHI/2019_games.html'

page = requests.get(URL)

# Get the contents of the page
soup = BeautifulSoup(page.content, 'html.parser')
print(soup)



```

```python
# Parsing the webpage - goal is to get as close to the data as possible and work your way down to the exact results

# Can find elements in the page by id, type of element (table, div) or a combination of both
results = soup.find(id='games')
#print(results)

results1 = soup.find('table')
#print(results1)

# this is to specific because the one table header is nested into the table and then we need to pull info from the table
results2 = soup.find('th')
#print(results2)
```

```python
# Iterating through data on a webpage

# find all the table rows inside the table
game_table = soup.find('table').find_all('tr')

# need a list to append everything to
all_stats = []

for row in game_table:
    # now we are iterating through all data but need specific data into that row
    
    # Indicate what stats we are scraping by 'data-stat' element 
    stats = ['date_game', 'game_start_time', 'network', 'opp_name', 'game_result', 'overtimes', 'pts', 'opp_pts', 'wins', 'losses', 'game_streak']
    
    # loop through all the table elements and the stats we are scraping amd put them in a list
    stats_list = [[td.getText() for td in row.findAll('td', {'data-stat': stat})] for stat in stats]
    
    all_stats.append(stats_list)
    
print(all_stats)
```

```python
# Data cleaning to put into dataframe

# we notice above the first list element is empty so get rid of it
all_stats = all_stats[1:]

# Notice how everything is in a column so we need to clean that
df = pd.DataFrame(all_stats)
df.columns = ['date_game', 'game_start_time', 'network', 'opp_name', 'game_result', 'overtimes', 'pts', 'opp_pts', 'wins', 'losses', 'game_streak']
df.head()
```
